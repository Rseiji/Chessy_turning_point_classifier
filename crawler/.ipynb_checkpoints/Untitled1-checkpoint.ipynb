```python
import os
import re
import sys
import pandas as pd


if __name__ == "__main__":
    all_files = sys.argv[1]
    new_files = sys.argv[2]

    downloaded_files = os.listdir('game_pgn_records/')
    downloaded_pgn = [int(re.findall('pgn_(\d+).txt', x)[0]) for x in downloaded_files]

    all_files = pd.read_csv(f'game_links/{all_files}', sep=';')

    print(f'Downloaded: {len(downloaded_pgn)}')
    print(f'Total: {len(all_files)}')

    absent_files = all_files[~all_files['game_id'].isin(downloaded_pgn)]
    absent_files.to_csv(f'game_links/{new_files}', sep=';', index=False)
```
